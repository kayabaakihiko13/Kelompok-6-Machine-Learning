```python
def pearson_corr(x, y):
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n
    S_xy = sum([(x[i]-mean_x)*(y[i]-mean_y) for i in range(n)])
    S_xx = sum([(x[i]-mean_x)**2 for i in range(n)])
    S_yy = sum([(y[i]-mean_y)**2 for i in range(n)])
    r = S_xy / (S_xx ** 0.5 * S_yy ** 0.5)
    return r
pearson_corr(x,y)
```
